4. **Как устроен формат JSON и чем он отличается от обычного текста?**
   JSON — это **структурированный текст**: словари, списки, числа, строки.
   Обычный текст — просто строки, без структуры.
   Пример JSON:

   ```json
   {"name": "Alex", "age": 25}
   ```

5. **Как использовать модуль `json` для чтения и записи структурированных данных?**

   * `json.dump()` — записать в файл (from python data types in json)
   * `json.load()` — считать из файла (from json in python data types)

     Пример:

   ```python
   import json
   with open("data.json", "w") as f:
       json.dump({"a": 1}, f)
   ```




- Чтение JSON-файла:
Для чтения JSON-файлов используется метод `load()`, считывающий весь файл целиком и возвращающий объект стандартного типа данных Python:


````python
import json
from pprint import pprint
with open("data.json", encoding="UTF-8") as file_in:
    records = json.load(file_in)
pprint(records)
````

````python
import json
from pprint import pprint

with open("data.json", encoding="UTF-8") as file_in:
    records = json.load(file_in) # load file from json to python data types
records[1]["group_number"] = 2   # change what need to change

with open("data.json", "w", encoding="UTF-8") as file_out:
    json.dump(records, file_out, ensure_ascii=False, indent=2) # then dump python data types to json
````

- `Что такое pprint?` (pretty-print) — это модуль в Python для форматирования и вывода сложных структур данных (списки, словари и т.д.) в более читаемом виде. Он улучшает читаемость, добавляя отступы и переносы строк.

````python
def factorial(n):
    # базовый случай — факториал от 0 равен 1
    if n == 0:
        return 1
    # рекурсивный случай: n! = n * (n-1)!
    return n * factorial(n - 1)

# print(factorial(0)) # 1
print(factorial(4))

# factorial(4)
# → 4 * factorial(3)        # не знаю factorial(3) равен сколько, подожду результат этого
# → 4 * (3 * factorial(2))  # та же идея, жду...
# → 4 * (3 * (2 * factorial(1)))
# → 4 * (3 * (2 * (1 * factorial(0))))
# → 4 * (3 * (2 * (1 * 1)))  ← здесь дошли до базовый случай: factorial(0) = 1
# → 4 * 3 * 2 * 1 = 24
````

Объект — это конкретный экземпляр класса, обладающий собственными значениями атрибутов. Объект знает, к какому классу он принадлежит, и может выполнять описанные в нём действия.

Создай класс Point, представляющий точку на плоскости (x, y).

метод __str__, который красиво выводит координаты ("Point(2, 3)")

метод __add__, чтобы можно было складывать две точки (Point(1,2) + Point(3,4))

````python
class Point:

  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __add__(self, other):
    return Point(self.x + other.x, self.y + other.y)

  def __str__(self):
    return f"Point({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)
print(p1 + p2)
````
